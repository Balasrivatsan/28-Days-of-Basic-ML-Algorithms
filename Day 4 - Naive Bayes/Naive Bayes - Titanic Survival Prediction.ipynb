Mapping the Text Data to Binary Values

```python
sex_info = set(df['Sex'])
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1}).astype(int)
print(df.head(5))
```

Segregating the Dataset into X and Y

```python
X = df.drop('Survived', axis = 'columns')
X
```

```python
Y = df.Survived
Y
```

Finding and Removing NaN values from X

```python
X.Age = X.Age.fillna(X.Age.mean())
```

Splitting Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

Selecting the Model

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

Training the Model

```python
model.fit(X_train, y_train)
```

```python
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
```

Predicting whether the Person Survived or Not

```python
pclassno = int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Person's Gender(0 - female, 1 - male): "))
age = int(input("Enter Person's Age: "))
fare = float(input("Enter Person's Fare: "))
person_new = [[pclassno, gender, age, fare]]

result = model.predict(person_new)
print(result)

if result == 0:
    print("Person wouldn't have Survived")
else:
    print("Person would have Survived")
```

Predicting the Test Data

```python
y_pred = model.predict(X_test)
print(np.column_stack((y_pred,y_test)))
```

Evaluating Model's Accuracy

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
